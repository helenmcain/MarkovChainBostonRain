```python
weather_june_july = pd.DataFrame({'Start Date': ['2025-06-27', '2025-07-04', '2025-07-11',
                                                 '2025-07-18', '2025-7-25'],
                                  'Real Weather': ['Only Saturday',
                                                   'No rain', 'No rain',
                                                   'Only Sunday', 'Friday and Sunday']})
display(weather_june_july)
```

Above is a Dataframe representing the Friday_Saturday_Sunday rain patterns from June 27th to July 27th of 2025

```python
matrix = stochastic_matrix(df_oneyear['State'].tolist())
display(matrix)
matrix.to_csv('/content/drive/MyDrive/Code_files/oneyear_helen.csv')
```

```python
print('Matrix generated with data from June 24th 2024-2025 (one year)')
print('Observed weather pattern from June 27th-July 27th (five weeks)')
walk_probability(matrix, 5, "three_day", "whole", ['Only Sunday', 'Only Saturday',
                                                   'No rain', 'No rain',
                                                   'Only Sunday', 'Friday and Sunday'])
```

Using the calendar year from June 24th 2024-2025 yields a probability of 0% because there were no instances in which a weekend where it rained only on Saturday was followed by a weekend where it rained only on Sunday.
